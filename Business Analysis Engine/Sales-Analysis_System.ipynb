```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 1: Sample synthetic customer data (age, annual income in k$, spending score)
data = pd.DataFrame({
    'CustomerID': range(1, 11),
    'Age': [19, 21, 20, 23, 31, 45, 33, 35, 40, 55],
    'Annual Income (k$)': [15, 16, 17, 20, 40, 70, 55, 58, 60, 90],
    'Spending Score (1-100)': [39, 81, 6, 77, 40, 5, 60, 54, 47, 30]
})

# Step 2: Select features for clustering
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Step 3: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)

# Step 5: Visualize the clusters (Income vs Spending Score)
plt.figure(figsize=(8, 6))
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'],
            c=data['Cluster'], cmap='viridis', s=100, edgecolor='k')
plt.title('Customer Segments (Proof of Concept)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()

# Step 6: Show clustered data
print(data[['CustomerID', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Cluster']])
```
